## Translation

### Sampling of dataset

```python
sys_sample_df = df.iloc[::1756]
sys_sample_df.reset_index(drop=True, inplace=True)
spa_l = []
fre_l = []
ger_l = []
```

```python
sys_sample_df
```

### Translator object for each language

```python
spa_trans = GoogleTranslator(source='english', target='spanish')
fre_trans = GoogleTranslator(source='english', target='french')
ger_trans = GoogleTranslator(source='english', target='german')
```

### Translating to each language

```python
for i in (sys_sample_df.iloc[:, 0]):
    spa_l.append(spa_trans.translate(i))
    fre_l.append(fre_trans.translate(i))
    ger_l.append(ger_trans.translate(i))
```

### Mergeing dataframes

```python
df_trans =  pd.DataFrame({'Spanish':spa_l, 'French':fre_l, 'German':ger_l})
```

```python
translated_df = pd.concat([sys_sample_df.iloc[:, 0], df_trans], axis=1)
```
