## Importing

```python
from textblob import Word
from textblob import TextBlob
import pandas as pd
```

## Function to correcting the spelling

```python
def word_spell_check(word):
    word = Word(word)
    crtd_word = word.correct()
    return crtd_word

def sentence_spell_check(sentence):
    string = TextBlob(sentence)
    crtd_string = string.correct()
    return crtd_string
```

### Reading the dataset of mispelled words and sentences

```python
df = pd.read_csv('spell_mistake.csv')
```

### Correcting spelling

```python
df_crt = pd.DataFrame({'crt':[]})
```

```python
for each_line in df.iloc[:, 0]:
    l_el = len(each_line.split())
    df1 = pd.DataFrame({'crt':[str(sentence_spell_check(each_line)) if l_el>1 else word_spell_check(each_line)]})
    df_crt = df_crt.append(df1, ignore_index=True)
```

### Setting the pandas dataframe to display all the rows

```python
pd.set_option('display.max_rows', None)
```

### Correct Spelled Result

```python
pd.concat([df, df_crt], axis=1)
```
